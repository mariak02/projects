#**Task 3 - IRIS FLOWER CLASSIFICATION**
1. The Iris flower dataset consists of three species: setosa, versicolor,
and virginica. These species can be distinguished based on their
measurements. It is categorized by their respective species.

2. The objective is to train a machine learning model that can learn from
these measurements and accurately classify the Iris flowers into
their respective species, based on their sepal and petal
measurements.

iris.png

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
iris = pd.read_csv('/content/IRIS.csv')
```

```python
# top 5 rows
iris.head()
```

```python
# information of data
iris.info()
```

```python
# statistical information for numerical columns in DataFrame
iris.describe()
```

```python
# number of samples in each species

iris['species'].value_counts()
```

```python
# visualize the dataset
sns.pairplot(iris, hue='species')
```

```python
# define target (y) and features (X)
y = iris['species']
X = iris.drop(['species'],axis=1)
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=42)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=200)
```

```python
# train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# model accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# accuracy score
accuracy_score(y_test,y_pred)
```

Here we got accuracy score 1.0 means the model is **100%** accurate. And it has made correct predictions for all the data points in the dataset.

```python
print(classification_report(y_test,y_pred))
```
